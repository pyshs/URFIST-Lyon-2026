## Exécuter un code

````python
# Calcule et affiche la proportion de mots de plus de 4 lettres d'une phrase
nombre_min_lettres = 4 # créer une variable
sortie_fichier = True # créer une variable
print("Calcul de la proportion de mots avec 4 lettres ou plus.") # afficher un texte
phrase = input("Écrivez une phrase: ") # demander une entrée utilisateur pour créer une variable
liste_mots = phrase.split(" ") # on coupe l'entrée aux espaces
nombre_mots = len(liste_mots)
if len(phrase) > 0: # on teste que la phrase est plus grande que 0 (que l'utilisateur a rentré un texte)
    # cas où il y a un texte
    compteur = 0 # initialiser du compteur
    for i in liste_mots: # boucle sur les mots
        if len(i) >= nombre_min_lettres: # condition de seuil pour chaque mot
            compteur += 1 # incrémentation du compteur
    proportion = round(100 * compteur / nombre_mots, 2)
    informations = {
        "Total": compteur,
        "Proportion": proportion,
        "Phrase": phrase,
        "Seuil": nombre_min_lettres,
    }
    sortie = (
        f"Proportion de mots avec {nombre_min_lettres} lettres ou plus : {proportion}"
    )
    print(sortie)
    if sortie_fichier:
        with open("resultat.txt", "w") as f:
            f.write(str(informations))
else:
    # cas où il n'y a pas de texte
    print("Vous n'avez rien écrit.")
````
