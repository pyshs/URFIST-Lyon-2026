# Construire une fonction

```python
nombre_min_lettres = 4
sortie_fichier = True
print("Calcul de la proportion de mots avec 4 lettres ou plus.")
phrase = input("Écrivez une phrase: ")
liste_mots = phrase.split(" ")
nombre_mots = len(liste_mots)
if len(phrase) > 0:
    compteur = 0
    for i in liste_mots:
        if len(i) >= nombre_min_lettres:
            compteur += 1
    proportion = round(100 * compteur / nombre_mots, 2)
    informations = {
        "Total": compteur,
        "Proportion": proportion,
        "Phrase": phrase,
        "Seuil": nombre_min_lettres,
    }
    sortie = (
        f"Proportion de mots avec {nombre_min_lettres} lettres ou plus : {proportion}"
    )
    print(sortie)
    if sortie_fichier:
        with open("resultat.txt", "w") as f:
            f.write(str(informations))
else:
    print("Vous n'avez rien écrit.")
```

## Fonction de compte de mots

```python
def proportion_mots_seuil(phrase, nombre_min_lettres = 4, sortie_fichier = True):
    """
    Retourner le nombre de mots de plus de nombre_min_lettres lettres
    """
    liste_mots = phrase.split(" ")
    nombre_mots = len(liste_mots)
    if len(phrase) > 0:
        compteur = 0
        for i in liste_mots:
            if len(i) >= nombre_min_lettres:
                compteur += 1
        proportion = round(100 * compteur / nombre_mots, 2)
        informations = {
            "Total": compteur,
            "Proportion": proportion,
            "Phrase": phrase,
            "Seuil": nombre_min_lettres,
        }
        if sortie_fichier:
            with open("resultat.txt", "w") as f:
                f.write(str(informations))
        return informations
    else:
        print("Vous n'avez rien écrit.")
        return None
```

```python
proportion_mots_seuil("Ceci est un texte")
```

```python
proportion_mots_seuil("Coucou")
```

```python
# récupérer du texte
phrase = input("Entrer une phrase")

# appeler la fonction
resultat = proportion_mots_seuil(phrase, nombre_min_lettres=10)

# afficher le retour
sortie = f"Proportion de mots avec {resultat['Seuil']} lettres ou plus : {resultat['Proportion']}"
print(sortie)
```

Appliquer avec la lecture de fichier

```python
# récupérer du texte
with open("./message.txt","r") as f:
    texte = f.read()
    
# appeler la fonction
resultat = proportion_mots_seuil(texte)

# afficher le retour
print(resultat)
```
